```python
import requests
from lxml import html
import pandas as pd
from requests_html import HTMLSession
```

```python
def get_link(url):
    session = HTMLSession()
    r = session.get(url)
    # article link selector path
    sel = '#content > div:nth-child(1) > div.main-column.ui-helper-clearfix > div.news-list-column-wide > ul > li > div.body > a'
    link_list = []
    try:
        links = r.html.find(sel)
        for link in links:
            link = list(link.absolute_links)[0]
            link_list.append(link)
        #print('10 latest company news link\n',link_list[0:10])
        # return top 10 news link
        return link_list[0:10]
    except:
        return None
```

```python
def get_news(url):
    try:
        company_news = []
        r = requests.get(url).content 
        sel = html.fromstring(r)
        news_title = sel.xpath('//*[@id="content"]/div[1]/div[1]/div[1]/article/header/h1/text()')
        news_content = sel.xpath('//*[@id="content"]/div[1]/div[1]/div[1]/article/div/text()')
        news = news_title + news_content
        return news
    except:
        return None
```

```python
def save_company_news(common_url):
    for epic in company_list:
        url = common_url + epic
        link_list = get_link(url)
        for url in link_list:
            news = get_news(url)
            for text in news:
                with open (epic + '.txt','a',encoding='utf-8') as f:
                    f.write(str(text))
        print("Successfully save {0} company news\n ".format(epic))
```

```python
if __name__ == '__main__':
    company = pd.read_csv('FTSE_250.csv')
    company_list = company.Epic
    common_url = 'http://www.stockmarketwire.com/company-news/?epic='
    save_company_news(common_url)
```

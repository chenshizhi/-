```python
import requests
from lxml import html
from bs4 import BeautifulSoup
from requests_html import HTMLSession
```

```python
#constituents > tbody > tr:nth-child(1) > td:nth-child(1) > a
```

```python
def get_link_name(url):
    #get company wiki link
    session = HTMLSession()
    r = session.get(url)
    # specific link selector path
    selector = '#constituents > tbody > tr > td:nth-child(1) > a'
    links = r.html.find(selector)
    link_list = []
    #print('Company wikipedia page link:')
    for link in links:
        link = list(link.absolute_links)[0]
        link_list.append(link)
        #print(link)
    
    #company name
    req = requests.get(url).content 
    sel = html.fromstring(req)
    epic = sel.xpath('//*[@id="constituents"]/tbody/tr/td[2]/text()')
    name_list = []
    for name in epic:
        name=name.strip('\n')
        name_list.append(name)
        #print(name)
    return link_list,name_list
```

```python
url = 'https://en.wikipedia.org/wiki/FTSE_250_Index'
links,names = get_link_name(url)
```

```python
def wiki_text(url):
    result = requests.get(url)
    c = result.content
    soup = BeautifulSoup(c)

    page_text = ''
    page = soup.find("div", id='mw-content-text').findAll('p')
    for text in page:
        page_text += '\n' + ''.join(text.findAll(text = True))
    return page_text
```

```python
print(wiki_text(links[0]))
```

```python
def SaveFile(content,filename):
    f=open( filename + '.txt',"w",encoding="utf-8")
    f.write(content)
    f.close()
```

```python
for i in range(len(links)):
    SaveFile(wiki_text(links[i]),names[i])
    print('completed ' + str(i))
```
